```python
import numpy as np
from math import gcd
import math 

            

tt = ['.#.####..#.#...#...##..#.#.##.',
      '..#####.##..#..##....#..#...#.',
      '......#.......##.##.#....##..#',
      '..#..##..#.###.....#.#..###.#.',
      '..#..#..##..#.#.##..###.......',
      '...##....#.##.#.#..##.##.#...#',
      '.##...#.#.##..#.#........#.#..',
      '.##...##.##..#.#.##.#.#.#.##.#',
      '#..##....#...###.#..##.#...##.',
      '.###.###..##......#..#...###.#',
      '.#..#.####.#..#....#.##..#.#.#',
      '..#...#..#.#######....###.....',
      '####..#.#.#...##...##....#..##',
      '##..#.##.#.#..##.###.#.##.##..',
      '..#.........#.#.#.#.......#..#',
      '...##.#.....#.#.##........#..#',
      '##..###.....#.............#.##',
      '.#...#....#..####.#.#......##.',
      '..#..##..###...#.....#...##..#',
      '...####..#.#.##..#....#.#.....',
      '####.#####.#.#....#.#....##.#.',
      '#.#..#......#.........##..#.#.',
      '#....##.....#........#..##.##.',
      '.###.##...##..#.##.#.#...#.#.#',
      '##.###....##....#.#.....#.###.',
      '..#...#......#........####..#.',
      '#....#.###.##.#...#.#.#.#.....',
      '.........##....#...#.....#..##',
      '###....#.........#..#..#.#.#..',
      '##...#...###.#..#.###....#.##.']

puzzle_input = tt
```

```python

def str_to_int_map(ccc,DBG=True):
    
    yy = len(ccc[0])
    xx = len(ccc)
    if(DBG): print("xx,yy",xx,yy)

    map_int = np.zeros((xx,yy), dtype=int)
    
    for x in range(0,xx):
        for y in range(0,yy):
            if(DBG): print("x,y",x,y)

            if ccc[x][y]=='#':
                map_int[x][y]=1
            else:
                map_int[x][y]=0
            if(DBG): print("x,y,map_int[x][y]",x,y,map_int[x][y])

    return map_int

def count_asteroids(mmm,x0,y0,DBG=True):
    
    yy = len(mmm[0])
    xx = len(mmm)

    visible = np.full((xx,yy), -1)
    if(DBG): print(visible)

    for x in range(0,xx):
        for y in range(0,yy):
            if (x==x0) and (y==y0): continue
            if mmm[x][y]==0: continue
            if(DBG): print("x0,y0",x0,y0)
            if(DBG): print("x,y",x,y)
            dx = (x-x0)
            dy = (y-y0)
            pgcd = gcd(dx, dy)
            ddx = dx//pgcd
            ddy = dy//pgcd
            if(DBG): print("ddx,ddy",ddx,ddy)
            curx=x0+ddx
            cury=y0+ddy
            visible_flag=1
            if(DBG): print("curx,cury",curx,cury)
            while((curx>=0) and (curx<xx) and (cury>=0) and (cury<yy)):
                if ((mmm[curx][cury]==1) and (visible[curx][cury]!=0)):
                    visible[curx][cury]=visible_flag
                    visible_flag=0
                curx += ddx
                cury += ddy
                if(DBG): print("curx,cury",curx,cury)

    if(DBG): print(visible)
    count = np.count_nonzero(visible == 1)
    return count,visible

def remove(mmm, vvv):
    yy = len(mmm[0])
    xx = len(mmm)

    for x in range(0,xx):
        for y in range(0,yy):
            if vvv[x][y]==1:
                mmm[x][y]==0
    return mmm

def function(ccc, x0,y0,maxboom, DBG=True):

    yy = len(ccc[0])
    xx = len(ccc)

    mmm = str_to_int_map(ccc, DBG)
    if(DBG): print(mmm)

    boom = 0
    ast_left = mmm.copy()
    while(True):
        asteroids,visible = count_asteroids(ast_left,x0,y0,DBG)
        if (boom+asteroids>maxboom):
            if(DBG): print("boom, asteroids, maxboom",boom, asteroids, maxboom)
            break
        boom += asteroids
        if(DBG): print("boom",boom)
        if(DBG): print("ast_left",ast_left)
        if(DBG): print("visible",visible)
        ast_left = remove(ast_left,visible)
        if(DBG): print("ast_left_remove",ast_left)
    
    # rotate and count
    angles={}
    for x in range(0,xx):
        for y in range(0,yy):
            if(ast_left[x][y]==1):
                angle=-math.atan2(y-y0,x-x0)+math.pi/2
                angles[angle]=(x,y)

    if(DBG): 
        for key in sorted(angles):
            print("key,angles[key]",key,angles[key])
            
    for key in sorted(angles):
        boom += 1
        if (boom==maxboom):
            ret = [angles[key],boom]
            break
    
    if(DBG): print("ret",ret)

    return ret

```

```python
def test(cc, x0,y0,maxboom,expected=None):
    result = str(function(cc,x0,y0,maxboom,True))
    flag = (result == expected)
    if (expected==None):
        print(str(cc) + " -> "+ str(result))
    else:
        print(str(cc) + " -> "+str(result), " -> "+ str(flag))
```
