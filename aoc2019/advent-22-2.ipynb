```python
def create_deck(size):
    return np.arange(size, dtype=np.int8)


# deal into new stack
# deal with increment 64
# cut 4395


def deck_to_str(deck):
    ret = ""
    for card in deck:
        ret += " "
        ret += str(card)
    return ret.strip()


def new_stack(deck, DBG=True):
    if DBG:
        print("new stack")
    ret = np.flip(deck, axis=0)
    return ret


def increment(deck, inc, DBG=True):
    if DBG:
        print("inc", inc)
    nb_cards = len(deck)
    ret = np.zeros(nb_cards, np.int8)
    index = 0
    for card in deck:
        ret[index % nb_cards] = card
        index += inc
    return ret


def cut(deck, inc, DBG=True):
    if DBG:
        print("cut", inc)
    ret = np.concatenate([deck[inc:], deck[:inc]])
    return ret


def function(deck, ccc, DBG=True):

    for line in ccc.splitlines():
        if DBG:
            print("line", line)
        if line == "deal into new stack":
            deck = new_stack(deck, DBG)
        elif line.startswith("deal with increment "):
            inc = int(line[line.rfind(" ") :])
            deck = increment(deck, inc, DBG)
        elif line.startswith("cut "):
            inc = int(line[line.rfind(" ") :])
            deck = cut(deck, inc, DBG)
        else:
            print("WRONG INSTRUCTION")

    return deck
```

```python
# returns index of card that ends up at index_card_after after "new stack"
def reverse_new_stack(len_deck, index_card_after, DBG=True):
    ret = len_deck - index_card_after - 1
    return ret


# returns index of card that ends up at index_card_after after "increment inc"
def reverse_increment(len_deck, index_card_after, inc, DBG=True):
    if DBG:
        print("inc", inc)
    ret = 0
    for k in range(len_deck):
        if (k * len_deck + index_card_after) % inc == 0:
            ret = (k * len_deck + index_card_after) // inc
            break
    if DBG:
        print("ret", ret)
    return ret


# returns index of card that ends up at index_card_after after "cut int"
def reverse_cut(len_deck, index_card_after, inc, DBG=True):
    if DBG:
        print("cut", inc)
    ret = (index_card_after + inc) % len_deck
    return ret


def input_to_commands(ccc, DBG=True):
    ret = []
    for line in ccc.splitlines():
        if DBG:
            print("line", line)
        if line == "deal into new stack":
            ret.append(("STACK", 0))
        elif line.startswith("deal with increment "):
            inc = int(line[line.rfind(" ") :])
            ret.append(("INC", inc))
        elif line.startswith("cut "):
            inc = int(line[line.rfind(" ") :])
            ret.append(("CUT", inc))
        else:
            print("WRONG INSTRUCTION")
    return ret


def reverse_function(len_deck, index_card_after, ccc, DBG=True):
    ret = index_card_after
    for command in reversed(ccc):
        if DBG:
            print("line", line)
        if command[0] == "STACK":
            ret = reverse_new_stack(len_deck, ret, DBG)
        elif command[0] == "INC":
            inc = command[1]
            ret = reverse_increment(len_deck, ret, inc, DBG)
        elif command[0] == "CUT":
            inc = command[1]
            ret = reverse_cut(len_deck, ret, inc, DBG)
        else:
            print("WRONG INSTRUCTION")
    return ret
```
