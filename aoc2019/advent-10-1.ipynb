```python
import numpy as np
from math import gcd

# fmt: off

tt = ['.#.####..#.#...#...##..#.#.##.',
      '..#####.##..#..##....#..#...#.',
      '......#.......##.##.#....##..#',
      '..#..##..#.###.....#.#..###.#.',
      '..#..#..##..#.#.##..###.......',
      '...##....#.##.#.#..##.##.#...#',
      '.##...#.#.##..#.#........#.#..',
      '.##...##.##..#.#.##.#.#.#.##.#',
      '#..##....#...###.#..##.#...##.',
      '.###.###..##......#..#...###.#',
      '.#..#.####.#..#....#.##..#.#.#',
      '..#...#..#.#######....###.....',
      '####..#.#.#...##...##....#..##',
      '##..#.##.#.#..##.###.#.##.##..',
      '..#.........#.#.#.#.......#..#',
      '...##.#.....#.#.##........#..#',
      '##..###.....#.............#.##',
      '.#...#....#..####.#.#......##.',
      '..#..##..###...#.....#...##..#',
      '...####..#.#.##..#....#.#.....',
      '####.#####.#.#....#.#....##.#.',
      '#.#..#......#.........##..#.#.',
      '#....##.....#........#..##.##.',
      '.###.##...##..#.##.#.#...#.#.#',
      '##.###....##....#.#.....#.###.',
      '..#...#......#........####..#.',
      '#....#.###.##.#...#.#.#.#.....',
      '.........##....#...#.....#..##',
      '###....#.........#..#..#.#.#..',
      '##...#...###.#..#.###....#.##.']
# fmt: on

puzzle_input = tt
```

```python
def str_to_int_map(ccc, DBG=True):

    yy = len(ccc[0])
    xx = len(ccc)
    if DBG:
        print("xx,yy", xx, yy)

    map_int = np.zeros((xx, yy), dtype=int)

    for x in range(0, xx):
        for y in range(0, yy):
            if DBG:
                print("x,y", x, y)

            if ccc[x][y] == "#":
                map_int[x][y] = 1
            else:
                map_int[x][y] = 0
            if DBG:
                print("x,y,map_int[x][y]", x, y, map_int[x][y])

    return map_int


def count_asteroids(mmm, x0, y0, DBG=True):

    yy = len(mmm[0])
    xx = len(mmm)

    visible = np.full((xx, yy), -1)
    if DBG:
        print(visible)

    for x in range(0, xx):
        for y in range(0, yy):
            if (x == x0) and (y == y0):
                continue
            if mmm[x][y] == 0:
                continue
            if DBG:
                print("x0,y0", x0, y0)
            if DBG:
                print("x,y", x, y)
            dx = x - x0
            dy = y - y0
            pgcd = gcd(dx, dy)
            ddx = dx // pgcd
            ddy = dy // pgcd
            if DBG:
                print("ddx,ddy", ddx, ddy)
            curx = x0 + ddx
            cury = y0 + ddy
            visible_flag = 1
            if DBG:
                print("curx,cury", curx, cury)
            while (curx >= 0) and (curx < xx) and (cury >= 0) and (cury < yy):
                if (mmm[curx][cury] == 1) and (visible[curx][cury] != 0):
                    visible[curx][cury] = visible_flag
                    visible_flag = 0
                curx += ddx
                cury += ddy
                if DBG:
                    print("curx,cury", curx, cury)

    if DBG:
        print(visible)
    count = np.count_nonzero(visible == 1)
    return count


def function(ccc, DBG=True):

    yy = len(ccc[0])
    xx = len(ccc)

    mmm = str_to_int_map(ccc, DBG)
    if DBG:
        print(mmm)

    x_max = -1
    y_max = -1
    asteroid_max = -1

    for x in range(0, xx):
        for y in range(0, yy):
            if mmm[x][y] == 0:
                continue
            asteroids = count_asteroids(mmm, x, y, DBG)
            if DBG:
                print("**x,y,asteroids", x, y, asteroids)
            if asteroids >= asteroid_max:
                if DBG:
                    print("x,y,max", x, y, asteroids)
                asteroid_max = asteroids
                x_max = x
                y_max = y

    return [(x_max, y_max), asteroid_max]
```

```python
def test(cc=None, expected=None):
    result = str(function(cc, False))
    flag = result == expected
    if expected == None:
        print(str(cc) + " -> " + str(result))
    else:
        print(str(cc) + " -> " + str(result), " -> " + str(flag))
```
