```python
CURSOR_UP_ONE = "\x1b[1A"
ERASE_LINE = "\x1b[2K"


def delete_last_lines(n=1):
    for _ in range(n):
        print(CURSOR_UP_ONE + ERASE_LINE + CURSOR_UP_ONE)


#        sys.stdout.write(CURSOR_UP_ONE)
#        sys.stdout.write(ERASE_LINE)


def print_field(xyids, DBG=True):
    coords = xyids.keys()
    if DBG:
        print(xyids)
    x_min = min(coords, key=lambda t: t[0])[0] - 1
    x_max = max(coords, key=lambda t: t[0])[0] + 1
    y_min = min(coords, key=lambda t: t[1])[1] - 1
    y_max = max(coords, key=lambda t: t[1])[1] + 1

    if DBG:
        print(x_min, x_max, y_min, y_max)

    for yy in range(y_min, y_max + 1):
        ss = ""
        for xx in range(x_min, x_max + 1):
            if (xx, yy) in xyids:
                ss += str(xyids[(xx, yy)])
            else:
                ss += "#"
        print(ss)

    # delete_last_lines(y_max-y_min)
```

```python
def connect(xx, yy, kkk, g_all, field):
    k_c = (xx, yy)
    if k_c in field:  # and (field[k_c]==1):
        kkc = (xx, yy, field[k_c])
        g_all.add_node(kkc)
        g_all.add_edge(kkc, kkk)


def create_world(ccc, DBG=True):
    field = {}
    keys = {}
    doors = {}
    g_all = nx.Graph()
    level = 0
    in_out = ""
    x = -1
    y = -1

    for line in ccc.splitlines():
        y += 1
        x = -1
        for c in line:
            x += 1
            if c == " " or c == "#":
                continue
            else:
                field[(x, y)] = c

    if DBG:
        print(field)
    if DBG:
        print_field(field)

    for k in field.keys():
        (x, y) = k
        kkk = (x, y, field[k])
        # connect
        # left
        connect(x - 1, y, kkk, g_all, field)
        # right
        connect(x + 1, y, kkk, g_all, field)
        # up
        connect(x, y - 1, kkk, g_all, field)
        # down
        connect(x, y + 1, kkk, g_all, field)
        #
        if (field[k] >= "a") and (field[k] <= "z"):
            keys[field[k]] = kkk
        elif (field[k] >= "A") and (field[k] <= "Z"):
            doors[field[k]] = kkk
        elif field[k] == "@":
            entry = kkk

    if DBG:
        print("***FINAL GRAPH")
    if DBG:
        print("nodes", len(g_all.nodes), list(g_all.nodes))
    if DBG:
        print("edges", len(g_all.edges), list(g_all.edges))
    if DBG:
        print("#nodes", len(g_all.nodes))
    if DBG:
        print("#edges", len(g_all.edges))
    if DBG:
        print("keys", len(keys), keys)
    if DBG:
        print("doors", len(doors), doors)
    if DBG:
        print("***END FINAL GRAPH")
    if DBG:
        # nx.draw(g, with_labels=True, font_weight='bold')
        # nx.draw_shell(g, with_labels=True, font_weight='bold')
        # nx.draw_circular(g, with_labels=True, font_weight='bold')
        nx.draw_kamada_kawai(g_all, with_labels=True, font_weight="bold")
        # nx.draw_random(g, with_labels=True, font_weight='bold')
        # nx.draw_spectral(g, with_labels=True, font_weight='bold')
        # nx.draw_spring(g_all, with_labels=True, font_weight='bold')

    return g_all, field, keys, doors, entry
```

```python
import operator

CUR_MIN = np.iinfo(np.int32).max


def function(
    g_all,
    field,
    keys_all,
    doors_all,
    position,
    keys_taken,
    full_path,
    full_path_length,
    DBG=True,
):
    global CUR_MIN
    # hint = deep copy graph, remove nodes of doors that we don't have the key for

    # 0. status = position + list of keys that we have
    # list available keys sorted by dist
    # if no key left, we're done, count steps we used and backtrack
    # for k in available keys
    # go to k (closest ?)
    # take key and goto 0

    graph = copy.deepcopy(g_all)

    keys_left_to_get = set(keys_all) - keys_taken

    if DBG:
        print("CUR_MIN", CUR_MIN)
    if DBG:
        print("position", position)
    if DBG:
        print("keys_all", keys_all)
    if DBG:
        print("keys_taken", keys_taken)
    if DBG:
        print("keys_left_to_get", keys_left_to_get)
    if DBG:
        print("doors_all", doors_all)

    if len(keys_left_to_get) == 0:
        print("**got them all")
        print("full_path", full_path)
        print("full_path_length", full_path_length)
        CUR_MIN = full_path_length
        return full_path, full_path_length

    sp = {}

    # remove doors for which we have no key from network
    for k in keys_left_to_get:
        k_up = k.upper()
        if k_up in doors_all:
            graph.remove_node(doors_all[k_up])

    if DBG:
        print("graph nodes", len(graph.nodes), list(graph.nodes))
    if DBG:
        print("graph edges", len(graph.edges), list(graph.edges))

    for k in keys_left_to_get:
        if DBG:
            print("looking up path to key", k)
        try:
            sp[k] = nx.shortest_path(graph, position, keys_all[k])
        except (nx.NetworkXNoPath, nx.NodeNotFound):
            # dead end
            sp[k] = []
        if DBG:
            print("sp[k]", sp[k])

    # LOOP ON KEYS AND RECURSE
    position2 = {}
    keys_taken2 = {}
    full_path2 = {}
    full_path_length2 = {}

    # for k in sp.keys():
    for k, v in sorted(sp.items(), key=operator.itemgetter(1)):
        if sp[k] == []:
            full_path2[k] = copy.deepcopy(full_path)
            full_path_length2[k] = np.iinfo(np.int32).max
        else:
            delta_path = len(sp[k]) - 1

            position2[k] = copy.deepcopy(position)
            keys_taken2[k] = copy.deepcopy(keys_taken)
            full_path2[k] = copy.deepcopy(full_path)
            full_path_length2[k] = copy.deepcopy(full_path_length)

            keys_taken2[k].add(k)
            position2[k] = keys_all[k]
            full_path2[k].append(k)
            full_path_length2[k] += delta_path

            if full_path_length2[k] >= CUR_MIN:
                full_path2[k] = copy.deepcopy(full_path)
                full_path_length2[k] = np.iinfo(np.int32).max
            else:
                full_path2[k], full_path_length2[k] = function(
                    g_all,
                    field,
                    keys_all,
                    doors_all,
                    position2[k],
                    keys_taken2[k],
                    full_path2[k],
                    full_path_length2[k],
                    DBG,
                )

    # get minimum
    best_key = min(full_path_length2, key=full_path_length2.get)

    if DBG:
        print("best_key", best_key)
    if DBG:
        print("full_path2[best_key]", full_path2[best_key])
    if DBG:
        print("full_path_length2[best_key]", full_path_length2[best_key])

    return full_path2[best_key], full_path_length2[best_key]
```

```python
def test(cc=None, expected=None, DBG=True):
    global CUR_MIN
    CUR_MIN = np.iinfo(np.int32).max
    g_all, field, keys_all, doors_all, entry = create_world(cc, DBG)
    # function(g_all,field,keys_all,doors_all,position,keys_taken,full_path,full_path_length,DBG=True):
    position = entry
    keys_taken = set()
    full_path = []
    full_path_length = 0
    full_path, full_path_length = function(
        g_all,
        field,
        keys_all,
        doors_all,
        position,
        keys_taken,
        full_path,
        full_path_length,
        DBG,
    )
    result = str(full_path_length)
    expected = str(expected)
    flag = result == expected
    print(
        "***" + str(cc) + " -> " + str(result), " -> " + str(flag) + " vs " + expected
    )
    print(full_path)
```
