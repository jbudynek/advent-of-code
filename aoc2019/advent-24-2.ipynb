```python

def create_worlds(ccc):
    x=0
    y=0
    worlds = {}
    worlds[0] = {}
    for line in ccc.splitlines():
        x=0
        for c in line:
            if(c=='.'):
                worlds[0][(x,y)]=0
            elif(c=='#'):
                worlds[0][(x,y)]=1
            else: print("WRONG")
            x+=1
        y+=1
    return worlds

def create_empty_level():
    world = {}
    for yy in range(5):
        for xx in range(5): 
            world[(xx,yy)]=0
    return world

def print_field(worlds, DBG=True):
    for level in worlds.keys():
        world = worlds[level]
    
        print("Level",level)
        for yy in range(5):
            ss = ""
            for xx in range(5):
                if (xx,yy)==(2,2):
                    ss += "?"
                elif (xx,yy) in world:
                    if (world[(xx,yy)]==1):
                        ss += "#"
                    elif (world[(xx,yy)]==0):
                        ss += "."
                    else:
                        ss += "WRONG"
                else:
                    ss += " "
            print(ss)
        print(" ")

def count_bugs_around(worlds, level, xx, yy):
    count = 0
    world = worlds[level]
    if (level+1) in worlds:
        world_in = worlds[level+1]
    else: 
        world_in = create_empty_level()
    if (level-1) in worlds:
        world_out = worlds[level-1]
    else: 
        world_out = create_empty_level()
    
    if (xx,yy)==(0,0):
        count += world[(1,0)] # RIGHT
        count += world[(0,1)] # DOWN
        count += world_out[(2,1)] # UP
        count += world_out[(1,2)] # LEFT
    elif (xx,yy)==(0,1):
        count += world[(1,1)] # RIGHT
        count += world[(0,2)] # DOWN
        count += world[(0,0)] # UP
        count += world_out[(1,2)] # LEFT
    elif (xx,yy)==(0,2):
        count += world[(1,2)] # RIGHT
        count += world[(0,3)] # DOWN
        count += world[(0,1)] # UP
        count += world_out[(1,2)] # LEFT
    elif (xx,yy)==(0,3):
        count += world[(1,3)] # RIGHT
        count += world[(0,4)] # DOWN
        count += world[(0,2)] # UP
        count += world_out[(1,2)] # LEFT
    elif (xx,yy)==(0,4):
        count += world[(1,4)] # RIGHT
        count += world_out[(2,3)] # DOWN
        count += world[(0,3)] # UP
        count += world_out[(1,2)] # LEFT
    elif (xx,yy)==(1,0):
        count += world[(2,0)] # RIGHT
        count += world[(1,1)] # DOWN
        count += world_out[(2,1)] # UP
        count += world[(0,0)] # LEFT
    elif (xx,yy)==(1,1):
        count += world[(2,1)] # RIGHT
        count += world[(1,2)] # DOWN
        count += world[(1,0)] # UP
        count += world[(0,1)] # LEFT
    elif (xx,yy)==(1,2):
        count += world_in[(0,0)] # RIGHT
        count += world_in[(0,1)] # RIGHT
        count += world_in[(0,2)] # RIGHT
        count += world_in[(0,3)] # RIGHT
        count += world_in[(0,4)] # RIGHT
        count += world[(1,3)] # DOWN
        count += world[(1,1)] # UP
        count += world[(0,2)] # LEFT
    elif (xx,yy)==(1,3):
        count += world[(2,3)] # RIGHT
        count += world[(1,4)] # DOWN
        count += world[(1,2)] # UP
        count += world[(0,3)] # LEFT
    elif (xx,yy)==(1,4):
        count += world[(2,4)] # RIGHT
        count += world_out[(2,3)] # DOWN
        count += world[(1,3)] # UP
        count += world[(0,4)] # LEFT
    elif (xx,yy)==(2,0):
        count += world[(3,0)] # RIGHT
        count += world[(2,1)] # DOWN
        count += world_out[(2,1)] # UP
        count += world[(1,0)] # LEFT
    elif (xx,yy)==(2,1):
        count += world[(3,1)] # RIGHT
        count += world_in[(0,0)] # DOWN
        count += world_in[(1,0)] # DOWN
        count += world_in[(2,0)] # DOWN
        count += world_in[(3,0)] # DOWN
        count += world_in[(4,0)] # DOWN
        count += world[(2,0)] # UP
        count += world[(1,1)] # LEFT
    elif (xx,yy)==(2,2): count = 0
    elif (xx,yy)==(2,3):
        count += world[(3,3)] # RIGHT
        count += world[(2,4)] # DOWN
        count += world_in[(0,4)] # UP
        count += world_in[(1,4)] # UP
        count += world_in[(2,4)] # UP
        count += world_in[(3,4)] # UP
        count += world_in[(4,4)] # UP
        count += world[(1,3)] # LEFT
    elif (xx,yy)==(2,4):
        count += world[(3,4)] # RIGHT
        count += world_out[(2,3)] # DOWN
        count += world[(2,3)] # UP
        count += world[(1,4)] # LEFT
    elif (xx,yy)==(3,0):
        count += world[(4,0)] # RIGHT
        count += world[(3,1)] # DOWN
        count += world_out[(2,1)] # UP
        count += world[(2,0)] # LEFT
    elif (xx,yy)==(3,1):
        count += world[(4,1)] # RIGHT
        count += world[(3,2)] # DOWN
        count += world[(3,0)] # UP
        count += world[(2,1)] # LEFT
    elif (xx,yy)==(3,2):
        count += world[(4,2)] # RIGHT
        count += world[(3,3)] # DOWN
        count += world[(3,1)] # UP
        count += world_in[(4,0)] # LEFT
        count += world_in[(4,1)] # LEFT
        count += world_in[(4,2)] # LEFT
        count += world_in[(4,3)] # LEFT
        count += world_in[(4,4)] # LEFT
    elif (xx,yy)==(3,3):
        count += world[(4,3)] # RIGHT
        count += world[(3,4)] # DOWN
        count += world[(3,2)] # UP
        count += world[(2,3)] # LEFT
    elif (xx,yy)==(3,4):
        count += world[(4,4)] # RIGHT
        count += world_out[(2,3)] # DOWN
        count += world[(3,3)] # UP
        count += world[(2,4)] # LEFT
    elif (xx,yy)==(4,0):
        count += world_out[(3,2)] # RIGHT
        count += world[(4,1)] # DOWN
        count += world_out[(2,1)] # UP
        count += world[(3,0)] # LEFT
    elif (xx,yy)==(4,1):
        count += world_out[(3,2)] # RIGHT
        count += world[(4,2)] # DOWN
        count += world[(4,0)] # UP
        count += world[(3,1)] # LEFT
    elif (xx,yy)==(4,2):
        count += world_out[(3,2)] # RIGHT
        count += world[(4,3)] # DOWN
        count += world[(4,1)] # UP
        count += world[(3,2)] # LEFT
    elif (xx,yy)==(4,3):
        count += world_out[(3,2)] # RIGHT
        count += world[(4,4)] # DOWN
        count += world[(4,2)] # UP
        count += world[(3,3)] # LEFT
    elif (xx,yy)==(4,4):
        count += world_out[(3,2)] # RIGHT
        count += world_out[(2,3)] # DOWN
        count += world[(4,3)] # UP
        count += world[(3,4)] # LEFT


    return count

def has_bug(worlds, level, xx,yy):
    world = worlds[level]
    if (xx,yy) in world: return (world[(xx,yy)]==1)
    else: return False


def tick(worlds, DBG=True):
    
    new_worlds = {}
    
    keys = list(worlds.keys()).copy()
    for level in keys:
        world = worlds[level]
        ccc = count_bugs_in_one_level(world)
        if (ccc>0):
            if (level-1) not in worlds:
                worlds[level-1] = create_empty_level()
            if (level+1) not in worlds:
                worlds[level+1] = create_empty_level()
    
    for level in worlds.keys():
        world = worlds[level]
        if(DBG): print(world)

        #    A bug dies (becoming an empty space) unless there is exactly one bug adjacent to it.
        # An empty space becomes infested with a bug if exactly one or two bugs are adjacent to it.
        
        new_world = {}
        
        for yy in range(5):
            for xx in range(5):
                if ((xx==2) and (yy==2)): 
                    new_world[(xx,yy)]=0
                    continue
                cnt = count_bugs_around(worlds, level, xx, yy)
                bug = has_bug(worlds, level, xx, yy)
                if (bug and cnt!=1):
                    new_world[(xx,yy)]=0
                elif (not bug and (cnt==1 or cnt==2)):
                    new_world[(xx,yy)]=1
                else: new_world[(xx,yy)]=world[(xx,yy)]
        new_worlds[level] = new_world
                    
    return new_worlds
        
    
def count_bugs_in_one_level(world):
    count_total = 0
    for yy in range(5):
        for xx in range(5):
            if ((xx==2) and (yy==2)): continue
            count_total += world[(xx,yy)]                    
    return count_total

def count_all_bugs(worlds, DBG=True):
    count_total = 0
    for level in worlds.keys():
        for yy in range(5):
            for xx in range(5):
                if ((xx==2) and (yy==2)): continue
                bug = has_bug(worlds, level, xx, yy)
                if (bug): count_total += 1
                    
    return count_total
        
    

```

```python
def test(cc, duration, DBG=True):
    new_worlds = create_worlds(cc)
    print_field(new_worlds)
    print("START WORLDS")

    for minute in range(duration):
        worlds = new_worlds
        new_worlds = tick(worlds,False)
        print_field(new_worlds,False)
        count = count_all_bugs(new_worlds,False)
        print("minute,count",minute,count)

```
