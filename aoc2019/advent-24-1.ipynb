```python
import numpy as np

puzzle_input = """#####
...##
#..#.
#....
#...#"""
```

```python
def create_world(ccc):
    x = 0
    y = 0
    world = {}
    for line in ccc.splitlines():
        x = 0
        for c in line:
            if c == ".":
                world[(x, y)] = 0
            elif c == "#":
                world[(x, y)] = 1
            else:
                print("WRONG")
            x += 1
        y += 1
    return world


def print_field(xyids, DBG=True):
    coords = xyids.keys()
    if DBG:
        print(xyids)
    x_min = min(coords, key=lambda t: t[0])[0]
    x_max = max(coords, key=lambda t: t[0])[0]
    y_min = min(coords, key=lambda t: t[1])[1]
    y_max = max(coords, key=lambda t: t[1])[1]

    if DBG:
        print(x_min, x_max, y_min, y_max)

    for yy in range(y_min, y_max + 1):
        ss = ""
        for xx in range(x_min, x_max + 1):
            if (xx, yy) in xyids:
                ss += str(xyids[(xx, yy)])
            else:
                ss += " "
        print(ss)


def function(ccc, DBG=True):

    return ccc


def count_bugs(world, xx, yy):
    count = 0
    if (xx - 1, yy) in world:
        count += world[(xx - 1, yy)]
    if (xx + 1, yy) in world:
        count += world[(xx + 1, yy)]
    if (xx, yy - 1) in world:
        count += world[(xx, yy - 1)]
    if (xx, yy + 1) in world:
        count += world[(xx, yy + 1)]
    return count


def has_bug(world, xx, yy):
    if (xx, yy) in world:
        return world[(xx, yy)] == 1
    else:
        return False


def tick(world, DBG=True):
    coords = world.keys()
    if DBG:
        print(world)
    x_min = min(coords, key=lambda t: t[0])[0]
    x_max = max(coords, key=lambda t: t[0])[0]
    y_min = min(coords, key=lambda t: t[1])[1]
    y_max = max(coords, key=lambda t: t[1])[1]

    if DBG:
        print(x_min, x_max, y_min, y_max)

    #    A bug dies (becoming an empty space) unless there is exactly one bug adjacent to it.
    # An empty space becomes infested with a bug if exactly one or two bugs are adjacent to it.

    new_world = {}

    for yy in range(y_min, y_max + 1):
        for xx in range(x_min, x_max + 1):
            cnt = count_bugs(world, xx, yy)
            bug = has_bug(world, xx, yy)
            if bug and cnt != 1:
                new_world[(xx, yy)] = 0
            elif not bug and (cnt == 1 or cnt == 2):
                new_world[(xx, yy)] = 1
            else:
                new_world[(xx, yy)] = world[(xx, yy)]
    return new_world


def score(world, DBG=True):
    coords = world.keys()
    if DBG:
        print(world)
    x_min = min(coords, key=lambda t: t[0])[0]
    x_max = max(coords, key=lambda t: t[0])[0]
    y_min = min(coords, key=lambda t: t[1])[1]
    y_max = max(coords, key=lambda t: t[1])[1]

    if DBG:
        print(x_min, x_max, y_min, y_max)

    #    A bug dies (becoming an empty space) unless there is exactly one bug adjacent to it.
    # An empty space becomes infested with a bug if exactly one or two bugs are adjacent to it.

    new_world = {}

    n = -1
    score = 0
    for yy in range(y_min, y_max + 1):
        for xx in range(x_min, x_max + 1):
            n += 1
            if has_bug(world, xx, yy):
                score += np.power(2, n)
    return score


def same(w1, w2, DBG=True):
    coords = w1.keys()
    if DBG:
        print(w1, w2)
    x_min = min(coords, key=lambda t: t[0])[0]
    x_max = max(coords, key=lambda t: t[0])[0]
    y_min = min(coords, key=lambda t: t[1])[1]
    y_max = max(coords, key=lambda t: t[1])[1]

    if DBG:
        print(x_min, x_max, y_min, y_max)

    for yy in range(y_min, y_max + 1):
        for xx in range(x_min, x_max + 1):
            if w1[(xx, yy)] != w2[(xx, yy)]:
                return False
    return True
```

```python
def test(cc=None, expected=None, DBG=True):
    new_world = create_world(cc)
    print_field(new_world)
    print("START WORLD")

    old_worlds = []
    for minute in range(100):
        world = new_world
        new_world = tick(world, False)
        print_field(new_world, False)
        biodiv = score(new_world, False)
        print("minute,biodiv", minute, biodiv)
        for ww in old_worlds:
            if same(new_world, ww, False):
                print("SAME WORLD")
                raise
        old_worlds.append(new_world.copy())
```
