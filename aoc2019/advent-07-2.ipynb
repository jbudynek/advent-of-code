```python

# ABCDE
#DE - two-digit opcode,      02 == opcode 2
# C - mode of 1st parameter,  0 == position mode
# B - mode of 2nd parameter,  1 == immediate mode
# A - mode of 3rd parameter,  0 == position mode,

def decode(c): # returns mode_a,mode_b,mode_c,opcode
    code_5="{:05d}".format(c)
    return int(code_5[0]),int(code_5[1]),int(code_5[2]),10*int(code_5[3])+int(code_5[4])

# returns OUTPUT,code,i,input_signals,input_index)
def function(code,i, input_signals,input_index,):
    #i=0
    #input_index=0
    
    mode_a,mode_b,mode_c,opcode = decode(code[i])
    
    
    while(True):

        #print(code[i],mode_a,mode_b,mode_c,opcode)
        
        if mode_c==0 and i+1<len(code):
            i1=code[i+1]
        else:
            i1=i+1
        if mode_b==0 and i+2<len(code):
            i2=code[i+2]
        else:
            i2=i+2
        if mode_a==0 and i+3<len(code):
            i3=code[i+3]
        else:
            i3=i+3


        if (opcode==99):
            print("HALT")       
            return ('HALT',code,i,input_signals,input_index)
        elif (opcode== 1): # addition
            code[i3] = code[i1]+code[i2]
            i = i +4
        elif (opcode == 2): # multiplication
            code[i3] = code[i1]*code[i2]
            i = i +4
        elif (opcode == 3): # input
            code[i1] = input_signals[input_index]
            input_index += 1
            i = i+2
        elif (opcode == 4): # output
            OUTPUT = code[i1]
            print(OUTPUT)
            i = i+2
            return (OUTPUT,code,i,input_signals,input_index)
        elif (opcode == 5): #jump-if-true
            if code[i1]!=0:
                i = code[i2]
            else:
                i=i+3
        elif (opcode == 6): #jump-if-false
            if code[i1]==0:
                i = code[i2]
            else:
                i=i+3
        elif (opcode == 7): #less than
            if code[i1]<code[i2]:
                code[i3]=1
            else:
                code[i3]=0
            i = i+4
        elif (opcode == 8): #equals
            if code[i1]==code[i2]:
                code[i3]=1
            else:
                code[i3]=0
            i = i+4
                
        mode_a,mode_b,mode_c,opcode = decode(code[i])

    # we never get there
    return "NEVER"
```

```python
import itertools

def find_best_sequence(program):
    max_thrust = 0
    best_seq = None
    perms = list(itertools.permutations([5,6,7,8,9]))
    
    for pp in perms:
        #print(pp)
        input_signal0 = 0
        output_signal = 0
        progs = list(range(5))
        i_s = list(range(5))
        input_signals = list(range(5))
        input_indexs = list(range(5))
        # first loop
        for prog in range(0,5): # a b c d e
            # start program
            progs[prog] = program.copy()
            i_s[prog] = 0
            input_signals[prog] = [pp[prog],input_signal0]
            input_indexs[prog] = 0
            #print(input_signals)
            # OUTPUT,code,i,input_signals,input_index
            output_signal,progs[prog],i_s[prog],input_signals[prog],input_indexs[prog] = function(progs[prog],i_s[prog],input_signals[prog],input_indexs[prog])
            input_signal0 = output_signal

        
        while(True):
            for prog in range(0,5): # a b c d e
                input_signals[prog] = [input_signal0]
                input_indexs[prog] = 0
                # OUTPUT,code,i,input_signals,input_index
                output_signal,progs[prog],i_s[prog],input_signals[prog],input_indexs[prog] = function(progs[prog],i_s[prog],input_signals[prog],input_indexs[prog])
                input_signal0 = output_signal
                
            if (output_signal=='HALT'):
                break
            else:
                latest_output_e = output_signal
            
        if (latest_output_e > max_thrust):
            max_thrust = latest_output_e
            best_seq = pp
    return best_seq, max_thrust
        
        
```

```python
def test(cc):
    print(str(cc) + " -> "+str(find_best_sequence(cc)))
```
