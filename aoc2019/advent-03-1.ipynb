```python

def xy2key(x,y):
    return str(x)+"_"+str(y)

def inc(di,x,y):
    kk = xy2key(x,y)
    if (not kk in di):
        di[kk] = 0
    di[kk] = di[kk]+1

def function(inp1, inp2):
    
    my_dict={}
    
    in1 = inp1.split(sep=',')
    in2 = inp2.split(sep=',')

    x0=0
    y0=0
    curx=x0
    cury=y0
    for dd in in1:
        #print(curx,cury,dd)
        dist = int(dd[1:])
        direction = dd[0]
        if direction=='R':
            for i in range(dist):
                inc(my_dict,curx+i,cury)
            curx = curx+dist
        if direction=='L':
            for i in range(dist):
                inc(my_dict,curx-i,cury)
            curx = curx-dist
        if direction=='U':
            for i in range(dist):
                inc(my_dict,curx,cury+i)
            cury = cury+dist
        if direction=='D':
            for i in range(dist):
                inc(my_dict,curx,cury-i)
            cury = cury-dist

    curx=x0
    cury=y0
    
    sol = 10000

    for dd in in2:
        dist = int(dd[1:])
        direction = dd[0]
        if direction=='R':
            for i in range(dist):
                inc(my_dict,curx+i,cury)
                kk = xy2key(curx+i,cury)
                if(my_dict[kk]==2 and (curx+i!=x0 and cury!=y0)): 
                    print(curx+i,cury)
                    sol = min(sol,abs(curx+i-x0) + abs(cury-y0))
                    print(sol)
            curx = curx+dist
        if direction=='L':
            for i in range(dist):
                inc(my_dict,curx-i,cury)
                kk = xy2key(curx-i,cury)
                if(my_dict[kk]==2 and (curx-i!=x0 and cury!=y0)): 
                    print(curx-i,cury)
                    sol = min(sol,abs(curx-i-x0) + abs(cury-y0))
                    print(sol)
            curx = curx-dist
        if direction=='U':
            for i in range(dist):
                inc(my_dict,curx,cury+i)
                kk = xy2key(curx,cury+i)
                if(my_dict[kk]==2 and (curx!=x0 and cury+i!=y0)): 
                    print(curx,cury+i)
                    sol = min(sol,abs(curx-x0) + abs(cury+i-y0))
                    print(sol)
            cury = cury+dist
        if direction=='D':
            for i in range(dist):
                inc(my_dict,curx,cury-i)
                kk = xy2key(curx,cury-i)
                if(my_dict[kk]==2 and (curx!=x0 and cury-i!=y0)): 
                    print(curx,cury-i)
                    sol = min(sol,abs(curx-x0) + abs(cury-i-y0))
                    print(sol)
            cury = cury-dist
          
    return sol

```
