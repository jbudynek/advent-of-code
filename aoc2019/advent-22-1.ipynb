```python

def create_deck(size):
    return np.arange(size)

#deal into new stack
#deal with increment 64
#cut 4395

def deck_to_str(deck):
    ret = ""
    for card in deck:
        ret +=" "
        ret += str(card)
    return ret.strip()

def new_stack(deck, DBG=True):
    if(DBG): print("new stack")
    ret = np.flip(deck,axis=0)
    return ret

def increment(deck, inc, DBG=True):
    if(DBG): print("inc", inc)
    nb_cards = len(deck)
    ret = np.zeros(nb_cards, int)
    index=0
    for card in deck:
        ret[index%nb_cards] = card
        index += inc
    return ret

def cut(deck, inc, DBG=True):
    if(DBG): print("cut", inc)
    ret = np.concatenate([deck[inc:] , deck[:inc]])
    return ret

def function(deck, ccc, DBG=True):
    
    for line in ccc.splitlines():
        if(DBG): print("line",line)
        if (line== "deal into new stack"):
            deck = new_stack(deck,DBG)
        elif (line.startswith("deal with increment ")):
            inc = int(line[line.rfind(" "):])
            deck = increment(deck, inc,DBG)
        elif (line.startswith("cut ")):
            inc = int(line[line.rfind(" "):])
            deck = cut(deck, inc,DBG)
        else: print("WRONG INSTRUCTION")
            
    
    return deck

```

```python
def test(cc=None, expected=None, DBG=True):
    deck = create_deck(10)
    result = function(deck, cc, DBG)
    result = deck_to_str(result)
    expected = str(expected)
    flag = (result == expected)
    print("***"+str(cc) + " -> "+str(result), " -> "+ str(flag))
```
